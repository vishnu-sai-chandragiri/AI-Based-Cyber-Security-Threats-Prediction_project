### Example 2: Control Flow (if-else and for loop)

This program demonstrates control flow by checking if a number is even or odd using an `if-else` statement and then calculates the sum of numbers from 1 to 5 using a `for` loop.

```c
#include <stdio.h>

int main() {
    // Demonstrate if-else statement: check if a number is even or odd
    int number = 7;
    printf("\nChecking if %d is even or odd:\n", number);
    if (number % 2 == 0) {
        printf("%d is an even number.\n", number);
    } else {
        printf("%d is an odd number.\n", number);
    }

    // Demonstrate for loop: calculate the sum of numbers from 1 to 5
    int sum = 0;
    int i;
    printf("\nCalculating sum of numbers from 1 to 5 using a for loop:\n");
    for (i = 1; i <= 5; i++) {
        sum += i; // Add current value of i to sum
    }
    printf("The sum of numbers from 1 to 5 is: %d\n", sum);

    return 0;
}
```

### Example 4: Pointers

This program illustrates the basic usage of pointers. It declares an integer variable, declares a pointer to that variable, and then prints both the value of the variable and the value accessed through the pointer. It also demonstrates how to change the variable's value using the pointer.

```c
#include <stdio.h>

int main() {
    // Declare an integer variable
    int value = 10;

    // Declare a pointer to an integer and make it point to 'value'
    int *ptr_to_value = &value;

    printf("\nDemonstrating Pointers:\n");
    printf("Original value: %d\n", value);
    printf("Value through pointer: %d\n", *ptr_to_value); // Dereferencing the pointer
    printf("Address of value: %p\n", (void*)&value);
    printf("Address stored in pointer: %p\n", (void*)ptr_to_value);

    // Change the value of 'value' using the pointer
    *ptr_to_value = 20;

    printf("\nValue after changing through pointer: %d\n", value);
    printf("Value through pointer after change: %d\n", *ptr_to_value);

    return 0;
}
```

````python
print("\nExample 4: Classes and Objects")

# Define a simple class
class Dog:
    # Constructor method with attributes
    def __init__(self, name, breed, age):
        self.name = name
        self.breed = breed
        self.age = age

    # Method to bark
    def bark(self):
        return f"{self.name} says Woof!"

    # Method to display dog's information
    def get_info(self):
        return f"Name: {self.name}, Breed: {self.breed}, Age: {self.age} years old."

# Create objects (instances) of the Dog class
dog1 = Dog("Buddy", "Golden Retriever", 3)
dog2 = Dog("Lucy", "Beagle", 5)

# Access attributes and call methods for dog1
print(f"Dog 1 Info: {dog1.get_info()}")
print(dog1.bark())

# Access attributes and call methods for dog2
print(f"Dog 2 Info: {dog2.get_info()}")
print(dog2.bark())
````

## Example 2: Handling User Input with Django Forms

This example extends the basic To-Do list application by adding functionality to create new To-Do items using Django Forms. It demonstrates how to define a form, render it in a template, handle form submissions, validate input, and save the data to the database.

#### Core Django Concepts Demonstrated:
*   **Django Forms**: Creating `forms.Form` or `forms.ModelForm` to handle user input.
*   **ModelForm**: A convenient way to create forms directly from a model.
*   **Form Rendering**: Displaying forms in templates with `{{ form.as_p }}`.
*   **Form Validation**: Automatic validation of submitted data.
*   **Saving Data**: Persisting valid form data to the database.
*   **Redirects**: Using `redirect()` after successful form submission to prevent resubmission.

#### Setup Instructions (Continuing from Example 1):
Assuming you have the `myproject` project and `todo` app set up from Example 1.

1.  **Define a Form:**
    Create a new file `todo/forms.py` and define a `ModelForm` for the `TodoItem` model:
    ```python
    # todo/forms.py

    from django import forms
    from .models import TodoItem

    class TodoItemForm(forms.ModelForm):
        class Meta:
            model = TodoItem
            fields = ['title', 'completed']
    ```

2.  **Update Views:**
    Modify `todo/views.py` to include a new view for adding items and update the `todo_list` view to display the form:
    ```python
    # todo/views.py

    from django.shortcuts import render, redirect
    from .models import TodoItem
    from .forms import TodoItemForm # Import the new form

    def todo_list(request):
        items = TodoItem.objects.all().order_by('-created_at')
        form = TodoItemForm() # Instantiate an empty form for display
        context = {
            'todo_items': items,
            'form': form # Pass the form to the template
        }
        return render(request, 'todo/todo/todo_list.html', context)

    def add_todo_item(request):
        if request.method == 'POST':
            form = TodoItemForm(request.POST) # Bind submitted data to the form
            if form.is_valid():
                form.save() # Save the new TodoItem to the database
                return redirect('todo_list') # Redirect to the list view
        else:
            form = TodoItemForm() # Display an empty form for GET requests
        # If request is GET or form is invalid, render the form again
        return render(request, 'todo/todo/add_todo.html', {'form': form})
    ```
